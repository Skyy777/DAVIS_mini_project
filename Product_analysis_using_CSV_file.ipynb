# Product Rating Analysis
## Mini Project using CSV Dataset

This project analyzes product ratings using a real-world styled CSV dataset
to understand customer satisfaction and product performance.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("product_ratings.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Average rating
df['Rating'].mean()
```

```python
# Highest rated product
df.loc[df['Rating'].idxmax()]
```

```python
# Lowest rated product
df.loc[df['Rating'].idxmin()]
```

```python
plt.figure()
sns.barplot(x='Product_Name', y='Rating', data=df)
plt.xticks(rotation=90)
plt.title("Product Ratings")
plt.show()
```

```python
plt.figure()
plt.hist(df['Rating'], bins=5)
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.title("Rating Distribution")
plt.show()
```

```python
category_avg = df.groupby('Category')['Rating'].mean().reset_index()

plt.figure()
sns.barplot(x='Category', y='Rating', data=category_avg)
plt.title("Average Rating by Category")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Review_Count'], df['Rating'])
plt.xlabel("Review Count")
plt.ylabel("Rating")
plt.title("Review Count vs Rating")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Price'], df['Rating'])
plt.xlabel("Price ()")
plt.ylabel("Rating")
plt.title("Price vs Rating")
plt.show()
```

```python
top5 = df.sort_values(by='Rating', ascending=False).head(5)

plt.figure()
sns.barplot(x='Product_Name', y='Rating', data=top5)
plt.xticks(rotation=45)
plt.title("Top 5 Highest Rated Products")
plt.show()
```

```python
category_count = df['Category'].value_counts()

plt.figure()
plt.pie(category_count, labels=category_count.index, autopct='%1.1f%%')
plt.title("Category Distribution")
plt.show()
```

### Observations
- Most products have ratings above 4, indicating good customer satisfaction.
- Electronics category contains the maximum number of products.
- Products with higher review counts appear more trustworthy.
- Expensive products generally receive better ratings.


### Conclusion
The analysis of the CSV-based product rating dataset provided valuable insights
into customer preferences, category performance, and pricing impact.
Such analysis helps businesses improve product quality and customer satisfaction.
