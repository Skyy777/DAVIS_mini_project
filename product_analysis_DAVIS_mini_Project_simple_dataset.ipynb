# Product Rating Analysis
## Mini Project for Data Analysis

### Introduction
Product ratings and reviews are important indicators of customer satisfaction.
This project analyzes product rating data to identify trends, top-rated products,
category-wise performance, and customer engagement using data visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: better visualization style
sns.set(style="whitegrid")
```

```python
data = {
    'Product_ID': ['P101', 'P102', 'P103', 'P104', 'P105'],
    'Product_Name': ['Smartphone', 'Headphones', 'Mixer Grinder', 'Shoes', 'Backpack'],
    'Category': ['Electronics', 'Electronics', 'Home Appliances', 'Fashion', 'Fashion'],
    'Rating': [4.5, 4.2, 3.8, 4.7, 3.5],
    'Review_Count': [120, 85, 60, 150, 40]
}

df = pd.DataFrame(data)
df
```


### Dataset Overview
The dataset contains product details, customer ratings, and review counts.


```python
# Average rating
df['Rating'].mean()
```

```python
# Highest rated product
df.loc[df['Rating'].idxmax()]
```

```python
# Lowest rated product
df.loc[df['Rating'].idxmin()]
```

```python
plt.figure()
sns.barplot(x='Product_Name', y='Rating', data=df)
plt.title("Product Ratings")
plt.xticks(rotation=45)
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure()
plt.hist(df['Rating'], bins=5)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.show()
```

```python
category_rating = df.groupby('Category')['Rating'].mean().reset_index()

plt.figure()
sns.barplot(x='Category', y='Rating', data=category_rating)
plt.title("Average Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Review_Count'], df['Rating'])
plt.title("Review Count vs Product Rating")
plt.xlabel("Number of Reviews")
plt.ylabel("Rating")
plt.show()
```

```python
top_products = df.sort_values(by='Rating', ascending=False).head(3)

plt.figure()
sns.barplot(x='Product_Name', y='Rating', data=top_products)
plt.title("Top 3 Highest Rated Products")
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.show()
```

```python
category_count = df['Category'].value_counts()

plt.figure()
plt.pie(category_count, labels=category_count.index, autopct='%1.1f%%')
plt.title("Product Category Distribution")
plt.show()
```

### Observations
- Most products have ratings above 4, indicating good customer satisfaction.
- Fashion products show higher ratings compared to other categories.
- Products with more reviews provide more reliable rating insights.
- Shoes is the highest-rated product in the dataset.


### Conclusion
This project demonstrates how product rating data can be analyzed using Python.
Visualization techniques help understand customer preferences, category performance,
